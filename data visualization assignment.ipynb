```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X_reduced = PCA(n_components=3).fit_transform(iris.data)
```

```python
Y_target = iris.target
```

```python
fig = plt.figure(1, figsize=(10,10))
ax = Axes3D(fig, elev=-150, azim=99)
ax.scatter(X_reduced[:,0], X_reduced[:,1], X_reduced[:, 2], c=Y_target, cmap=plt.cm.rainbow_r)

ax.set_title("PCA = 3 on Iris dataset")
ax.set_xlabel("first vector")
ax.set_ylabel("second vecor")
ax.set_zlabel("Third vector")

plt.show()
```
